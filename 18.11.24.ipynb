```python
import numpy as np
import pandas as pd
```

```python
mobile_df = pd.read_csv(r"C:\Users\123\Downloads\mobile.csv")
```

```python
print(mobile_df)
```

```python
mobile_df['first_name'] = mobile_df['Customer Name'].apply(lambda x: x.split()[0])
```

```python
print(mobile_df)
```

```python
mobile_df['Cellolar_company'] = mobile_df['Phone Number'].apply(lambda c: 'Pelephone' if c[:3] == '050' else 'Golan' if c[:3] == '051' else 'Cellcom' if c[:3] == '052' else 'Motorola' if c[:3] == '053' else 'Partner' if c[:3] == '054' else None)

print(mobile_df)
```

```python
print(mobile_df['Cellolar_company'].value_counts())
print(f"Mobile companies are: {mobile_df['Cellolar_company'].unique()}")
print(f"There are {mobile_df['Cellolar_company'].nunique()} Mobile companies")
```

```python
mobile_df['Kind_of_customer/By lambda'] = mobile_df.apply(lambda row: 'Talker' if row['Talked Minutes'] > row['Internet Minutes'] else 'Surfer',axis=1)
print(mobile_df)
```

```python
def talker_or_surfer(Talked_Minutes , Internet_Minutes):
    return "Talker" if Talked_Minutes>Internet_Minutes else "Surfer"

mobile_df['Kind_of_customer/By vectorize'] = np.vectorize(talker_or_surfer)(mobile_df['Talked Minutes'],mobile_df['Internet Minutes'])
print(mobile_df)
```
